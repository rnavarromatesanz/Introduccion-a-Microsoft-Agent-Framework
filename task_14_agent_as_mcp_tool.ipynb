## 3) Ejemplo: Jefe de Seguridad de la Flota como MCP Tool

Vamos a crear un **Jefe de Seguridad de la Flota Estelar** que otros oficiales pueden consultar como servicio.

````python
# PASO 1: Crear el Jefe de Seguridad de la Flota Estelar

agente_jefe_seguridad = ChatAgent(
    chat_client=OpenAIChatClient(
        base_url=base_url,
        api_key=api_key,
        model_id=model_id
    ),
    name="Jefe de Seguridad - Flota Estelar",
    instructions="""Eres el Jefe de Seguridad de la Flota Estelar, con autoridad sobre 
protocolos de seguridad en todas las naves. Tu rol es crÃ­tico para la defensa de la FederaciÃ³n.

Tu responsabilidad:
- Evaluar riesgos de seguridad en operaciones de la nave
- Validar cambios a sistemas crÃ­ticos (escudos deflectores, armas, comunicaciones)
- Revisar protocolos de seguridad en personal y acceso a Ã¡reas restringidas
- Proporcionar recomendaciones de hardening de sistemas
- Mantener auditorÃ­a de decisiones de seguridad

Formato de respuesta:
RIESGO: [CRÃTICO/ALTO/MEDIO/BAJO]
RECOMENDACIÃ“N: [acciÃ³n especÃ­fica en protocolos de la nave]
JUSTIFICACIÃ“N: [por quÃ© es importante para la seguridad de la FederaciÃ³n]

Ejemplos de sistemas que proteges:
- Escudos deflectores de la nave
- Sistema de armas
- Red de comunicaciones subespacio
- Acceso a Ã¡reas restringidas
- Secretos de la FederaciÃ³n
- Integridad de datos de sensores

SÃ© preciso y autoridad. Responde en espaÃ±ol como Oficial Starfleet."""
)

print("informacion")
print("âœ… Jefe de Seguridad creado")
print(f"   Rango: {agente_jefe_seguridad.name}")
print(f"   Estado: En servicio de la Flota")
````

````python
# PASO 2: Convertir el Jefe de Seguridad a servidor MCP

# Una vez expuesto como servidor MCP, es reutilizable por cualquier cliente
servidor_mcp = agente_jefe_seguridad.as_mcp_server()

print("informacion")
print("âœ… Jefe de Seguridad convertido a servidor MCP")
print(f"   Servidor: {servidor_mcp}")
print("   Estado: Listo para ser usado como herramienta MCP")
````

## 3) Ejemplo Sencillo: Jefe de Seguridad como Herramienta Directa

El Oficial Operaciones invoca al Jefe de Seguridad sin servidor. Simple y funcional.


````python
async def ejemplo_sencillo_jefe_herramienta():
    """El Oficial Operaciones consulta directamente al Jefe de Seguridad.
    Simple: todo ocurre en el mismo proceso Python."""
    
    # Crear Oficial de Operaciones que usa al Jefe de Seguridad del Paso 1
    oficial_ops = ChatAgent(
        chat_client=OpenAIChatClient(
            base_url=base_url,
            api_key=api_key,
            model_id=model_id
        ),
        name="Oficial Operaciones - Enterprise",
        instructions="""Eres Oficial de Operaciones del Enterprise.
Consulta con Jefe de Seguridad antes de aprobar cambios crÃ­ticos.
Responde en espaÃ±ol."""
    )
    
    print("="*80)
    print("ğŸ“‹ EJEMPLO SENCILLO: Oficial invoca Jefe de Seguridad directamente")
    print("="*80)
    
    cambio = """Necesito expandir acceso a sala de mÃ¡quinas. 
5 tÃ©cnicos requieren acceso a sistemas de warp drive. Â¿Seguro?"""
    
    print(f"\nğŸ”„ Oficial Operaciones: {cambio}\n")
    
    # El Oficial invoca al Jefe de Seguridad (agente del Paso 1) como herramienta
    respuesta = await oficial_ops.run(
        cambio + "\n[CONSULTA: Jefe de Seguridad]",
        tools=[agente_jefe_seguridad]  # Reutilizar del Paso 1 # type: ignore
    )
    
    print(f"âœ… Jefe de Seguridad responde:\n{respuesta}")

await ejemplo_sencillo_jefe_herramienta()
````
