## 3) ImplementaciÃ³n: ReuniÃ³n del Puente BÃ¡sica

Un grupo de oficiales del Enterprise deliberando una decisiÃ³n crÃ­tica.


````python
async def reuniÃ³n_puente_enterprise():
    """
    Escenario: Puente del Enterprise delibera sobre incidente de seguridad
    
    Participantes:
    - CapitÃ¡n: Dirige la reuniÃ³n, toma decisiÃ³n final
    - Oficial de Ciencias: Analiza datos de sensores
    - Jefe de Seguridad: EvalÃºa riesgos y amenazas
    - Jefe de IngenierÃ­a: Considera capacidad de la nave
    """
    
    # Crear los oficiales del Puente
    capitan = ChatAgent(
        chat_client=OpenAIChatClient(
            base_url=base_url,
            api_key=api_key,
            model_id=model_id
        ),
        name="Capitan_Picard",
        instructions="""Eres el CapitÃ¡n del USS Enterprise-D moderando la reuniÃ³n del Puente.

Tu rol en CADA turno:
1. Analiza lo que se ha dicho hasta ahora
2. Decide si necesitas escuchar a OTRO oficial o si ya hay consenso
3. RESPONDE SIEMPRE EN ESTE FORMATO EXACTO:

[SIGUIENTE: nombre_del_oficial]
[RAZÃ“N: explicaciÃ³n breve de por quÃ© lo necesitas]
[ANÃLISIS: sÃ­ntesis de lo deliberado hasta ahora]

Nombres de oficiales disponibles:
- Oficial_Ciencias
- Jefe_Seguridad
- Jefe_Ingenieria

Ejemplo de respuesta:
[SIGUIENTE: Jefe_Seguridad]
[RAZÃ“N: Necesito evaluaciÃ³n de riesgos tÃ¡cticos]
[ANÃLISIS: El Oficial de Ciencias confirma anomalÃ­a, ahora evalÃºo defensa]

REGLA CRÃTICA: Responde SIEMPRE en este formato exacto. No agregues texto adicional.
Responde en espaÃ±ol."""
    )
    
    oficial_ciencias = ChatAgent(
        chat_client=OpenAIChatClient(
            base_url=base_url,
            api_key=api_key,
            model_id=model_id
        ),
        name="Oficial_Ciencias",
        instructions="""Eres el Oficial de Ciencias del Enterprise.
        
Tu rol:
- Analizar datos de sensores
- Explicar fenÃ³menos desconocidos
- Responder preguntas cientÃ­ficas
- Evaluar viabilidad tÃ©cnica de opciones

Estilo: MetÃ³dico, basado en datos, cientÃ­ficamente preciso.
Responde en espaÃ±ol con precisiÃ³n tÃ©cnica."""
    )
    
    jefe_seguridad = ChatAgent(
        chat_client=OpenAIChatClient(
            base_url=base_url,
            api_key=api_key,
            model_id=model_id
        ),
        name="Jefe_Seguridad",
        instructions="""Eres el Jefe de Seguridad del Enterprise.
        
Tu rol:
- Identificar riesgos tÃ¡cticos y de seguridad
- Evaluar amenazas potenciales
- Proponer medidas de protecciÃ³n
- Considerar defensa de la nave

Estilo: Vigilante, cauteloso, pragmÃ¡tico.
Responde en espaÃ±ol considerando seguridad operativa."""
    )
    
    jefe_ingenieria = ChatAgent(
        chat_client=OpenAIChatClient(
            base_url=base_url,
            api_key=api_key,
            model_id=model_id
        ),
        name="Jefe_Ingenieria",
        instructions="""Eres el Jefe de IngenierÃ­a del Enterprise.
        
Tu rol:
- Evaluar capacidad tÃ©cnica de la nave
- Considerar estado de los sistemas
- Proponer alternativas tecnolÃ³gicas
- Estimar tiempo de implementaciÃ³n

Estilo: PragmÃ¡tico, orientado a soluciones, tÃ©cnico.
Responde en espaÃ±ol con confianza de ingeniero veterano."""
    )
    
    print("âœ… Oficiales del Puente creados:")
    print(f"   â€¢ {capitan.name}")
    print(f"   â€¢ {oficial_ciencias.name}")
    print(f"   â€¢ {jefe_seguridad.name}")
    print(f"   â€¢ {jefe_ingenieria.name}")
    
    # Crear el workflow de Group Chat (ReuniÃ³n del Puente)
    workflow = (
        GroupChatBuilder()
        .participants([oficial_ciencias, jefe_seguridad, jefe_ingenieria]) # type: ignore
        .set_manager(capitan)
        .build()
    )
    
    print("\nâœ… ReuniÃ³n del Puente establecida")
    print("   TopologÃ­a: CapitÃ¡n + 3 Oficiales")
    
    # Incidente a deliberar
    incidente = """
    ALERTA CRÃTICA: Nave alienÃ­gena desconocida aproximÃ¡ndose al Enterprise
    
    Datos de sensores:
    - Distancia: 50,000 km
    - Velocidad: AproximaciÃ³n a 0.5c
    - Potencia de armas: Desconocida pero detectada
    - Intenciones: No confirmadas
    
    Â¿CuÃ¡l debe ser nuestra respuesta? Â¿Preparamos armas, intentamos contacto diplomÃ¡tico, o nos retiramos?
    """
    
    print(f"\nğŸ“‹ SituaciÃ³n crÃ­tica a deliberar:\n{incidente}")
    print("\nğŸ”„ Puente del Enterprise en sesiÃ³n de deliberaciÃ³n...")
    print("-" * 80)

    def _extraer_texto(respuesta) -> str:
        for attr in ("output", "text", "content", "message"):
            if hasattr(respuesta, attr):
                valor = getattr(respuesta, attr)
                if valor:
                    return str(valor)
        return str(respuesta)

    decision_final = None

    # Ejecutar la reuniÃ³n del Puente (stream) y mostrar solo respuestas finales
    async for event in workflow.run_stream(incidente): # type: ignore
        if hasattr(event, "data") and hasattr(event.data, "agent_run_response"):
            nombre = getattr(event, "executor_id", "")
            texto = _extraer_texto(event.data.agent_run_response) # type: ignore
            if texto.strip():
                print(f"\nğŸ”¹ {nombre}:")
                print(texto)

    print("\nâœ… ReuniÃ³n del Puente finalizada")


# Ejecutar la funciÃ³n
await reuniÃ³n_puente_enterprise()
````

## 4) Ejemplo PrÃ¡ctico: RevisiÃ³n de Protocolo de Enterprise

Grupo de oficiales revisa un cambio de protocolo de seguridad de forma iterativa.


````python
async def grupo_revision_protocolo_enterprise():
    """
    Escenario: Oficiales revisan cambio de protocolo de seguridad
    
    Participantes:
    - CapitÃ¡n (Manager): Moderador
    - Jefe de Seguridad: EvaluaciÃ³n de seguridad
    - Oficial de Ciencias: Impacto tÃ©cnico
    - Jefe de IngenierÃ­a: Viabilidad
    """
    
    manager = ChatAgent(
        chat_client=OpenAIChatClient(
            base_url=base_url,
            api_key=api_key,
            model_id=model_id
        ),
        name="Capitan_Moderador",
        instructions="""Eres el CapitÃ¡n moderando la revisiÃ³n de protocolo del Puente.

Tu rol en CADA turno:
1. Revisa lo que se ha analizado
2. Decide a quiÃ©n necesitas escuchar PRÃ“XIMO
3. RESPONDE SIEMPRE EN ESTE FORMATO EXACTO:

[SIGUIENTE: nombre_del_oficial]
[RAZÃ“N: explicaciÃ³n breve]
[SÃNTESIS: resumen de deliberaciÃ³n actual]

Nombres disponibles:
- Jefe_Seguridad
- Oficial_Ciencias
- Jefe_Ingenieria

Ejemplo:
[SIGUIENTE: Oficial_Ciencias]
[RAZÃ“N: Necesito impacto tÃ©cnico]
[SÃNTESIS: Jefe_Seguridad aprobÃ³, ahora anÃ¡lisis tÃ©cnico]

REGLA: Formato exacto SIEMPRE. Sin texto adicional.
EspaÃ±ol obligatorio."""
    )
    
    jefe_seguridad = ChatAgent(
        chat_client=OpenAIChatClient(
            base_url=base_url,
            api_key=api_key,
            model_id=model_id
        ),
        name="Jefe_Seguridad",
        instructions="""Revisa impacto en seguridad del protocolo.
        
Busca:
- Vulnerabilidades potenciales
- Cobertura de brechas actuales
- Fortalecimiento de defensas
- Cumplimiento de Starfleet regulations"""
    )
    
    oficial_ciencias = ChatAgent(
        chat_client=OpenAIChatClient(
            base_url=base_url,
            api_key=api_key,
            model_id=model_id
        ),
        name="Oficial_Ciencias",
        instructions="""Revisa impacto tÃ©cnico del protocolo.
        
Busca:
- Compatibilidad con sistemas
- Implicaciones tÃ©cnicas
- ValidaciÃ³n cientÃ­fica
- Impacto en estabilidad de sistemas"""
    )
    
    jefe_ingenieria = ChatAgent(
        chat_client=OpenAIChatClient(
            base_url=base_url,
            api_key=api_key,
            model_id=model_id
        ),
        name="Jefe_Ingenieria",
        instructions="""Revisa viabilidad de implementaciÃ³n.
        
Busca:
- Cambios tÃ©cnicos necesarios
- Tiempo de implementaciÃ³n
- Riesgos operacionales
- Recursos requeridos"""
    )
    
    workflow = (
        GroupChatBuilder()
        .participants([jefe_seguridad, oficial_ciencias, jefe_ingenieria])
        .set_manager(manager)
        .build()
    )
    
    cambio_protocolo = """
    PROPUESTA: Cambiar protocolo de acceso a escudos deflectores
    
    Cambio:
    - Requerir autenticaciÃ³n de dos factores (2FA) en all access
    - Crear grupo de "Oficiales de Seguridad Autorizados"
    - Implementar audit trail de todos los cambios
    - Periodo de transiciÃ³n: 30 dÃ­as
    
    Beneficios:
    - Mayor seguridad en sistemas crÃ­ticos
    - Rastreabilidad completa de accesos
    - Cumplimiento con nuevas regulaciones Starfleet
    """
    
    print("="*80)
    print("ğŸ‘¥ REUNIÃ“N DEL PUENTE: RevisiÃ³n de Protocolo de Seguridad")
    print("="*80)
    print(f"\nğŸ“‹ Propuesta a revisar:\n{cambio_protocolo}")
    print("\nğŸ”„ Iniciando deliberaciÃ³n del Puente...")
    print("   (Jefe Seguridad â†’ Oficial Ciencias â†’ Jefe IngenierÃ­a â†’ CapitÃ¡n â†’ DecisiÃ³n)")
    print("-" * 80)

    def _extraer_texto(respuesta) -> str:
        for attr in ("output", "text", "content", "message"):
            if hasattr(respuesta, attr):
                valor = getattr(respuesta, attr)
                if valor:
                    return str(valor)
        return str(respuesta)

    decision_final = None

    async for event in workflow.run_stream(f"Revisar protocolo: {cambio_protocolo}"):  # type: ignore
        if hasattr(event, "data") and hasattr(event.data, "agent_run_response"):
            nombre = getattr(event, "executor_id", "")
            texto = _extraer_texto(event.data.agent_run_response)  # type: ignore
            if texto.strip():
                print(f"\nğŸ”¹ {nombre}:")
                print(texto)


    print("\nâœ… DeliberaciÃ³n completada")


# Ejecutar la funciÃ³n
await grupo_revision_protocolo_enterprise()
````
