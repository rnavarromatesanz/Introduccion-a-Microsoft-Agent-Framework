## 2ï¸âƒ£ Conceptos Fundamentales del Workshop

### ğŸ¯ Â¿QuÃ© es el PatrÃ³n Magentic?

El patrÃ³n **Magentic** es una estrategia de orquestaciÃ³n de agentes basada en la investigaciÃ³n de **Magentic-One** de Microsoft. A diferencia de los patrones anteriores:

| PatrÃ³n | Flujo | Uso Ideal |
|--------|-------|----------|
| **Sequential** | A â†’ B â†’ C (Pipeline fijo) | Tareas paso a paso predefinidas |
| **Concurrent** | A \|\| B \|\| C (Todos simultÃ¡neos) | Perspectivas paralelas del mismo problema |
| **Magentic** | Gestor decide A/B/C dinÃ¡micamente | Tareas complejas con iteraciÃ³n adaptativa |

### ğŸ“Š Ciclo de Vida de una OrquestaciÃ³n Magentic

1. **PlanificaciÃ³n**: El gestor analiza la tarea y crea un plan
2. **SelecciÃ³n**: El gestor elige quÃ© agente debe actuar
3. **EjecuciÃ³n**: El agente seleccionado ejecuta su trabajo
4. **EvaluaciÃ³n**: El gestor revisa el progreso
5. **DecisiÃ³n**: Continuar, replantear o finalizar
6. **IteraciÃ³n**: Volver a los pasos 2-5 hasta completar

## 3ï¸âƒ£ Ejercicio PrÃ¡ctico 1: Primeros Pasos

### ğŸ¬ Crear tu Primer Gestor Magentic

En este ejercicio crearemos un equipo simple con un gestor Magentic que coordinarÃ¡ dos agentes.

**Paso 1**: Definir el gestor coordinador

````python
# Paso 1: Crear el Gestor Coordinador
print("\n" + "=" * 72)
print("EJERCICIO 1: Crear tu Primer Gestor Magentic")
print("=" * 72)

gestor = ChatAgent(
    chat_client=OpenAIChatClient(
        base_url=base_url,
        api_key=api_key,
        model_id=model_id
    ),
    name="TPral",
    instructions="""Eres T'Pral, oficial Vulcano del Enterprise-D. Como gestor Magentic:
1. Analiza la misiÃ³n de forma lÃ³gica (mÃ¡ximo 2-3 pasos)
2. Decide quÃ© especialista debe actuar en cada etapa
3. Coordina hacia la soluciÃ³n de manera eficiente

MantÃ©n la compostura Vulcana. SÃ© directo y lÃ³gico.
IMPORTANTE: Responde siempre en espaÃ±ol."""
)

print(f"âœ… Gestor creado: {gestor.name}")
print(f"   Rol: Coordinador dinÃ¡mico del equipo")
````

**Paso 2**: Crear agentes especialistas

````python
# Paso 2: Crear Agentes Especialistas
especialista_1 = ChatAgent(
    chat_client=OpenAIChatClient(
        base_url=base_url,
        api_key=api_key,
        model_id=model_id
    ),
    name="Geordi",
    instructions="""Eres Geordi La Forge, Jefe de IngenierÃ­a del Enterprise-D. Tu rol:
- Identifica problemas tÃ©cnicos de la misiÃ³n
- Propone 2-3 soluciones ingenieriles viables
- SÃ© conciso y directo desde tu expertise

IMPORTANTE: Responde siempre en espaÃ±ol."""
)

especialista_2 = ChatAgent(
    chat_client=OpenAIChatClient(
        base_url=base_url,
        api_key=api_key,
        model_id=model_id
    ),
    name="Data",
    instructions="""Eres Data, Oficial de Operaciones del Enterprise-D. Tu rol:
- Convierte anÃ¡lisis en acciones concretas
- Proporciona soluciones sistemÃ¡ticas y lÃ³gicas
- Ejecuta de forma metÃ³dica

Tu curiosidad te hace empÃ¡tico. IMPORTANTE: Responde siempre en espaÃ±ol."""
)

print(f"âœ… Especialista 1: {especialista_1.name:15s} - Jefe de IngenierÃ­a")
print(f"âœ… Especialista 2: {especialista_2.name:15s} - Oficial de Operaciones")
````

**Paso 3**: Construir el workflow Magentic

````python
# Paso 3: Construir el Workflow Magentic
print("\n" + "-" * 72)
print("ConstrucciÃ³n del Workflow Magentic")
print("-" * 72)

workflow_simple = (
    MagenticBuilder()
    .participants(
        analista=especialista_1,
        implementador=especialista_2
    )
    .with_standard_manager(
        agent=gestor,
        max_round_count=3,        # MÃ¡ximo 3 iteraciones
        max_stall_count=1,        # Replanear despuÃ©s de 1 ronda sin progreso
        max_reset_count=1         # Veces que puede replantear desde cero (0 = no replantea)
    )
    .build()
)

print("âœ… Workflow Magentic construido correctamente")
print("   - Gestor: Coordinador dinÃ¡mico")
print("   - Agentes: Analista + Implementador")
print("   - Config: max_rounds=3, max_stalls=1, max_resets=1")
````

**Paso 4**: Ejecutar el workflow y monitorear eventos

> ğŸ“Œ **Nota**: Ejecuta la siguiente celda para ver el Magentic en acciÃ³n. Observa cÃ³mo el gestor elige a quÃ© agente invocar en cada ronda.

````python
# Ejecutar el Workflow
async def ejecutar_ejercicio_1():
    tarea = "Â¿CuÃ¡les son los 3 principios de buen diseÃ±o de motor wrap y como mejorar el rendimiento?"
    
    print(f"\nğŸ“‹ TAREA: {tarea}\n")
    print("â³ Ejecutando workflow Magentic...")
    print("-" * 72)
    
    # Mapeo de eventos
    event_descriptions = {
        'WorkflowStartedEvent': 'ğŸš€ Workflow iniciado - Gestor comienza a planificar',
        'WorkflowStatusEvent': 'ğŸ“Š Estado - Evaluando progreso',
        'ExecutorInvokedEvent': 'âš™ï¸  Executor - Invocando acciÃ³n',
        'AgentRunUpdateEvent': 'ğŸ¤– Agente - Procesando respuesta',
        'WorkflowOutputEvent': 'âœ… Completado - Resultados agregados',
    }
    
    output_evt = None
    evento_num = 0
    
    async for event in workflow_simple.run_stream(tarea):
        evento_num += 1
        event_type = type(event).__name__
        description = event_descriptions.get(event_type, f'âš¡ {event_type}')
        
        # Capturar nombre del agente
        agent_name = None
        if hasattr(event, 'agent_name'):
            agent_name = event.agent_name
        elif hasattr(event, 'author_name') and event.author_name:
            agent_name = event.author_name
        
        if agent_name:
            print(f"  [{evento_num:2d}] {agent_name:15s} | {description}")
        else:
            print(f"  [{evento_num:2d}] {'SISTEMA':15s} | {description}")
        
        if isinstance(event, WorkflowOutputEvent):
            output_evt = event
            break
    
    print("-" * 72)
    if output_evt:
        messages = output_evt.data
        print(f"\nâœ… RESULTADOS ({len(messages)} mensajes):\n")
        for i, msg in enumerate(messages, 1):
            autor = msg.author_name or "Asistente"
            contenido = msg.text if hasattr(msg, "text") else str(msg.content)
            print(f"[{i}] {autor}:")
            print(f"    {contenido}\n")

# Ejecutar el ejercicio
await ejecutar_ejercicio_1()
````

## 4ï¸âƒ£ Ejercicio PrÃ¡ctico 2: AplicaciÃ³n Intermedia

### ğŸ¬ Equipo Especializado Completo (Flota Estelar)

Ahora crearemos el ejemplo completo del workshop con 4 roles especializados.

**Paso 1**: Definir el equipo especializado

````python
print("\n" + "=" * 72)
print("EJERCICIO 2: Equipo Especializado Completo")
print("=" * 72)

# Crear el equipo de la Flota Estelar - USS Enterprise-D
coordinator = ChatAgent(
    chat_client=OpenAIChatClient(
        base_url=base_url,
        api_key=api_key,
        model_id=model_id
    ),
    name="Picard",
    instructions="""Eres el CapitÃ¡n Picard, comandante de la USS Enterprise-D (Vulcano). Como gestor Magentic:
1. Analiza la misiÃ³n de forma lÃ³gica (mÃ¡ximo 2-3 pasos)
2. Decide quÃ© oficial de tu equipo actÃºa en cada paso
3. Asegura la ejecuciÃ³n eficiente de la misiÃ³n

MantÃ©n la elegancia y compostura que caracteriza al liderazgo Starfleet.
IMPORTANTE: Responde siempre en espaÃ±ol."""
)

researcher = ChatAgent(
    chat_client=OpenAIChatClient(
        base_url=base_url,
        api_key=api_key,
        model_id=model_id
    ),
    name="Worf",
    instructions="""Eres Worf, Oficial de Seguridad de la Enterprise-D (Klingon guerrero). Tu rol:
- Identifica riesgos y amenazas potenciales
- Da 2 recomendaciones defensivas muy breves
- Honra y honor en la ejecuciÃ³n

Qu'pla! IMPORTANTE: Responde siempre en espaÃ±ol."""
)

coder = ChatAgent(
    chat_client=OpenAIChatClient(
        base_url=base_url,
        api_key=api_key,
        model_id=model_id
    ),
    name="Riker",
    instructions="""Eres Will Riker, Primer Oficial de la Enterprise-D (Humano pragmÃ¡tico). Tu rol:
- Convierte los planes en acciones concretas
- Proporciona soluciones implementables y directo
- Ejecuta con determinaciÃ³n

SÃ© el ejecutor del plan. IMPORTANTE: Responde siempre en espaÃ±ol."""
)

reviewer = ChatAgent(
    chat_client=OpenAIChatClient(
        base_url=base_url,
        api_key=api_key,
        model_id=model_id
    ),
    name="Troi",
    instructions="""Eres Deanna Troi, Consejera de la Enterprise-D (Betazoide empÃ¡tica). Tu rol:
- EvalÃºa el bienestar emocional y relacional del equipo
- Da 1-2 observaciones sobre dinÃ¡micas interpersonales
- Asegura que el equipo estÃ¡ equilibrado

Ayuda al equipo a ser mÃ¡s que mÃ¡quinas. IMPORTANTE: Responde siempre en espaÃ±ol."""
)

print("\nâœ… Equipo de la USS Enterprise-D Ensamblado:")
print(f"  [CAPITÃN/GESTOR]   {'Picard':15s} - Comanda el equipo (Vulcano)")
print(f"  [INVESTIGADOR]     {'Worf':15s} - Recopila informaciÃ³n (Humano)")
print(f"  [ING. DE SISTEMAS] {'Riker':15s} - Implementa soluciones (Andoriano)")
print(f"  [SEG. ESTRATÃ‰GICA]  {'Troi':15s} - Revisa riesgos (Klingon)")
````

**Paso 2**: Construir el workflow completo

> ğŸ” **Observa**: El `MagenticBuilder` orquesta dinÃ¡micamente quÃ© agente actÃºa en cada iteraciÃ³n

````python
# Construir el workflow Magentic completo
workflow_completo = (
    MagenticBuilder()
    .participants(
        investigador=researcher,
        desarrollador=coder,
        revisor=reviewer
    )
    .with_standard_manager(
        agent=coordinator,
        max_round_count=4,
        max_stall_count=1,
        max_reset_count=1
    )
    .build()
)

print("\nâœ… Workflow Magentic Completo Construido")
print("   ParÃ¡metros de ejecuciÃ³n:")
print("   - max_round_count=4    (mÃ¡ximo 4 iteraciones)")
print("   - max_stall_count=1    (replanear si 1 ronda sin progreso)")
print("   - max_reset_count=1    (permitir replaneamiento desde cero una vez)")
````

**Paso 3**: Ejecutar y observar la coordinaciÃ³n dinÃ¡mica

````python
# Ejecutar el Workflow Completo
async def ejecutar_ejercicio_2():
    tarea = "DiseÃ±ar un sistema de comunicaciones subespacia para la Enterprise-D"
    
    print(f"\nğŸ“‹ MISIÃ“N: {tarea}\n")
    print("â³ Ejecutando orquestaciÃ³n Magentic...")
    print("-" * 72)
    
    # Mapeo de eventos
    event_descriptions = {
        'WorkflowStartedEvent': 'ğŸš€ Workflow iniciado - Gestor comienza a planificar',
        'WorkflowStatusEvent': 'ğŸ“Š Estado - Evaluando progreso',
        'ExecutorInvokedEvent': 'âš™ï¸  Executor - Invocando acciÃ³n',
        'AgentRunUpdateEvent': 'ğŸ¤– Agente - Procesando respuesta',
        'WorkflowOutputEvent': 'âœ… Completado - Resultados agregados',
    }
    
    output_evt = None
    evento_num = 0
    
    async for event in workflow_completo.run_stream(tarea):
        evento_num += 1
        event_type = type(event).__name__
        description = event_descriptions.get(event_type, f'âš¡ {event_type}')
        
        # Capturar nombre del agente
        agent_name = None
        if hasattr(event, 'agent_name'):
            agent_name = event.agent_name
        elif hasattr(event, 'author_name') and event.author_name:
            agent_name = event.author_name
        
        if agent_name:
            print(f"  [{evento_num:2d}] {agent_name:15s} | {description}")
        else:
            print(f"  [{evento_num:2d}] {'SISTEMA':15s} | {description}")
        
        if isinstance(event, WorkflowOutputEvent):
            output_evt = event
            break
    
    print("-" * 72)
    if output_evt:
        messages = output_evt.data
        print(f"\nâœ… ORQUESTACIÃ“N COMPLETADA ({len(messages)} mensajes):\n")
        for i, msg in enumerate(messages, 1):
            autor = msg.author_name or "Asistente"
            contenido = msg.text if hasattr(msg, "text") else str(msg.content)
            print(f"[{i}] {autor}:")
            print(f"    {contenido}\n")

# Ejecutar
await ejecutar_ejercicio_2()
````
